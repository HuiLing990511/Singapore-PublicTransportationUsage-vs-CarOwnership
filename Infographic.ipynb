# Quartile Analysis for Income Data
Identify the boundaries of Q1, median (Q2), and Q3 for dimension reduction in income data.

```python
import re

def parse_bounds(interval):
    if "Below" in interval:
        return 0, 1000
    elif "andOver" in interval:
        return 20000, 22500  # estimated upper bound
    else:
        # Extract numbers using regex
        numbers = list(map(int, re.findall(r'\d{1,3}(?:_\d{3})*', interval)))
        # Remove underscores from formatted numbers (like 1_000)
        numbers = [int(str(n).replace('_', '')) for n in numbers]
        return numbers[0], numbers[1] + 1  # +1 to make upper bound exclusive

def compute_quartile(intervals, counts, quartile):
    N = sum(counts)
    target = quartile * N / 4

    cumulative = 0
    for i, count in enumerate(counts):
        prev_cumulative = cumulative
        cumulative += count
        if cumulative >= target:
            lower, upper = parse_bounds(intervals[i])
            width = upper - lower

            F = prev_cumulative
            f = count
            L = lower
            Q = L + ((target - F) / f) * width
            return Q
    return None

# Data
intervals = ['Below_1_000_Total', '1_000_1_999_Total', '2_000_2_999_Total',
             '3_000_3_999_Total', '4_000_4_999_Total', '5_000_5_999_Total',
             '6_000_6_999_Total', '7_000_7_999_Total', '8_000_8_999_Total',
             '9_000_9_999_Total', '10_000_10_999_Total', '11_000_11_999_Total',
             '12_000_12_999_Total', '13_000_13_999_Total', '14_000_14_999_Total',
             '15_000_17_499_Total', '17_500_19_999_Total', '20_000andOver_Total']

counts = [27203, 75725, 85571, 100056, 104126, 115691, 116525, 119565, 118512,
          117838, 109893, 101201, 92751, 88837, 78620, 171676, 125767, 427898]

# Compute quartiles
q1 = compute_quartile(intervals, counts, 1)
q2 = compute_quartile(intervals, counts, 2)
q3 = compute_quartile(intervals, counts, 3)

print(f"Q1: ${q1:,.2f}, Median (Q2): ${q2:,.2f}, Q3: ${q3:,.2f}")
```

## 1. Does gender affect choice of transportation between private car and public transport?

Interpretation of the Plot
Visualizing associations between:
1. Transport Types (Private, Public) → red points
2. Income Brackets (Low, Middle, High) → blue points

Axis: Component 1
- This is the first principal axis from CA — it captures the most significant dimension of association in the data.

How to interpret distances:

For example:
- Private ↔ High income → strong positive association.
- Public ↔ Low/Middle income → strong association too.

This kind of pattern makes intuitive sense: higher-income households tend to use private transport more often.
